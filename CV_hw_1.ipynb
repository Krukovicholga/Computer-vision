```python
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np
import imageio
import scipy.signal
```

```python
img = imageio.imread('rainbow.jpg')
```

```python
print(img.shape)
```

```python
plt.figure(figsize=(8, 8))
img = img.astype(np.float32) / 255.
plt.imshow(img)
```

Нелинейное увеличение яркости

```python
plt.figure(figsize=(8, 8))
plt.imshow(img*1.5)
```

Получение ЧБ изображения по цветному

```python
img_bw = np.mean(img, axis=2)
```

```python
plt.figure(figsize=(8, 8))
plt.imshow(img_bw, cmap='gray')
```

Свёртка цветного изобрежения

```python
win_size = (11, 11)
w = np.ones(win_size, dtype=np.float32)
w = w / np.sum(w)
out_img = []
for c in range(3):
    out_img.append(scipy.signal.correlate2d(img[:,:,c], w, mode='valid'))
out_img = np.stack(out_img, axis=2)
```

```python
plt.figure(figsize=(8, 8))
plt.imshow(out_img)
```
